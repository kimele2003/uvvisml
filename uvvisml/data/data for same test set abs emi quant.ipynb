```python
empty_idx = joung_df[joung_df['Absorption max (nm)'].isna()].index
joung_df.drop(index=empty_idx, inplace=True)
```

```python
empty_idx = joung_df[joung_df['Quantum yield'].isna()].index
joung_df.drop(index=empty_idx, inplace=True)
joung_df
```

```python
columns = {'Chromophore':'smiles','Solvent':'solvent', 'Absorption max (nm)':'abs_peakwavs_max', 
           'Emission max (nm)':'emi_peakwavs_max', 'abs FWHM (nm)':'abs_bandwidth',
           'emi FWHM (nm)':'emi_bandwidth', 'Quantum yield':'quantum_yield', 'Lifetime (ns)':'log_lifetime',
           'log(e/mol-1 dm3 cm-1)':'abs_molar_ext_coeff'}

joung_df.rename(columns=columns, 
                inplace=True)
joung_df
```

```python
joung_df.drop(axis=1, labels=['log_lifetime', 'abs_molar_ext_coeff', 'abs FWHM (cm-1)', 'emi FWHM (cm-1)', 'abs_bandwidth', 'emi_bandwidth', 'Molecular weight (g mol-1)', 'Reference'], inplace=True)
joung_df['source'] = 'deep4chem'
joung_df
```

```python
# # Filtering

# ### Remove Molecules that Cannot be Sanitized by RDKit

def sanitize_smiles(smiles):
    try:
        smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
    except:
        smiles = np.nan
    return smiles

joung_df['smiles'] = joung_df['smiles'].apply(lambda x: sanitize_smiles(x))
joung_df['solvent'] = joung_df['solvent'].apply(lambda x: sanitize_smiles(x))
joung_df.dropna(inplace=True)


# ### Remove Clusters (SMILES containing ".")

cluster_idx = joung_df[joung_df['smiles'].str.contains('\.')].index
#print('Removing {} rows'.format(len(cluster_idx)))
joung_df.drop(index=cluster_idx, inplace=True)


# # Export DataFrame to CSV File

x = ['smiles', 'solvent', 'abs_peakwavs_max', 'emi_peakwavs_max', 'quantum_yield', 'source']
joung_df[x].to_csv('~/uvvisml/uvvisml/data/processed/data_same_test_set_abs_emi_quant.csv', index=False)
```
